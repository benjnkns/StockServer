### 2. RSI ( Relative Strength Index )

```python
RSI_AAPL = []
RSI_OMC = []
```

```python
avg_gain = []
avg_loss = []
gains = []
losses = []
for i in range (14, len(data)):
    for j in range(i-14, i):
        if AAPL_Close[j+1]>AAPL_Close[j]:
            gains.append(AAPL_Close[j+1] - AAPL_Close[j])
        else:
            losses.append(AAPL_Close[j] - AAPL_Close[j+1])
    avg_gain.append(sum(gains)/len(gains))
    avg_loss.append(sum(losses)/len(losses))
    gains = []
    losses = []
len(gains), len(losses), len(avg_gain), len(avg_loss)
```

```python
for i in range (0,14):
    RSI_AAPL.append(0)
for i in range (0, len(data)-14):
    RSI_AAPL.append (100 - (100 / (1 + avg_gain[i] / avg_loss[i])))
len(RSI_AAPL)
```

```python
pyplot.plot(AAPL_Close[-200:])
pyplot.plot(RSI_AAPL[-200:])
pyplot.legend(['Close', 'RSI'])
pyplot.show()
```

```python
import numpy as np
from scipy.signal import butter,filtfilt
# Filter requirements.
T = 5        # Sample Period
fs = 30       # sample rate, Hz
cutoff = 5      # desired cutoff frequency of the filter, Hz ,      slightly higher than actual 1.2 Hz
nyq = 0.5 * fs  # Nyquist Frequency
order = 2       # sin wave can be approx represented as quadratic
n = int(T * fs) # total number of samples
```

```python
def butter_lowpass_filter(data, cutoff, fs, order):
    normal_cutoff = cutoff / nyq
    # Get the filter coefficients 
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    y = filtfilt(b, a, data)
    return y
```

```python
OMC_Filtered = butter_lowpass_filter(inversed_OMC[:,0], cutoff, fs, order)
pyplot.plot(OMC_Filtered[-50:])
pyplot.plot(OMC_Close[-50:, 0])
pyplot.legend(['predictions', 'Adj close'])
```

```python
print(metrics.mean_squared_error(OMC_Filtered[-50:], OMC_Close[-50:, 0]))
```
