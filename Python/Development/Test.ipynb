This is the test harness for my algorithms. It will simulate 10 years of running this prediction algorithm once a week with a starting wallet of $1000.

```python
import importlib
import import_ipynb
import datetime
import pandas
import yfinance as yf
from pandas_datareader import data as pdr
from datetime import timedelta

# MY STUFF
import Engine
importlib.reload(Engine)
from Engine import make_predictions

# Override yFinance with Pandas
yf.pdr_override() # <-- Here is the fix
```

```python
def run_simulation(Years, Cash, Tickers, Model):
    
    #define variables
    num_weeks = int(Years*365 / 7)
    
    for week in reversed(range(2, num_weeks)):
        
        #Predict closing prices for each week.
        date = datetime.date.today() - timedelta(weeks = week)
        print(date)
        predictions = make_predictions(Tickers, date, 10)

        #Calcualte the correct stock to purchase
        total_gains = []
        for i in range(0, len(Tickers)):
            total_gains.append(predictions[i][2] - predictions[i][1])
            print (Tickers[i] + "is expected to gain " + str(total_gains[i]))
            max_gain = max(total_gains)
            stock_to_purchase = Tickers[total_gains.index(max_gain)]
    
        #Purchase the stock.
        if (max_gain > 0):
            data = pdr.get_data_yahoo(stock_to_purchase, start = date, end = date + timedelta(days = 4))
            data = data.values
            purchase_price = data[0, 3]
            #check for NaN
            if (data[-1,3]==data[-1,3]):
                sell_price = data[-1, 3]
            else:
                sell_price = data[-2, 3]
            
            #Simulate purcahse and sell of the stock.
            stocks_purchased = int(Cash / purchase_price)
            Cash = Cash - stocks_purchased * purchase_price
            Cash = Cash + stocks_purchased * sell_price
            
        else:
            stocks_purchased = 0
        
        #Print Summary.
        print ("Purchased " + str(stocks_purchased) + " stocks of " + stock_to_purchase)
        print ("Cash = " + str(Cash))
    return Cash
```
